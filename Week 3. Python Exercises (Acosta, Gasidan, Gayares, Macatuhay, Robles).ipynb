###### 10. Write a program that will ask the user to input a number and then, the program will check and determine whether a given number is an Armstrong Number or not. The program also asks the user if the user wants repeat the program or not. Armstrong Number (Narcissistic Number) is a number that is the sum of its own digits each raised to the power of the number of digits 153 = 1^3 + 5^3 + 3^3

```python
def armstrong():
    print("\nARMSTRONG NUMBER")
    print()
    num = int(input("Input number: "))
    print()
    result=0

    temp = num
    while temp > 0:
        digit = temp % 10
        result += digit ** 3
        temp //= 10

    if num == result:
        print("The number ", num ," is an Armstrong number.")
    else:
        print("The number ", num ," is not an Armstrong number.")
    
    repeat=input("Do you want to continue ? (Y/N): ")
    if (repeat.upper() == "Y"):
        armstrong()
    else:
        print("\nEND OF PROGRAM")

if __name__ == '__main__':
    armstrong()
```
